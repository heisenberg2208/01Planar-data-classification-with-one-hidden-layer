```python
#defining number of nodes at each stage
def layer_sizes(X,Y):
    n_x = X.shape[0] #here X is 2x400
    n_y = Y.shape[0] #here Y is 1x400
    n_h = 4
    
    return (n_x,n_h,n_y)
```

```python
X_assess, Y_assess = layer_sizes_test_case()
(n_x , n_h , n_y) = layer_sizes(X_assess, Y_assess)
print(n_x)
print(n_h)
print(n_y)
```

```python
#initialise parameters
def initialize_parameters(n_x,n_h,n_y):
    np.random.seed(2)
    W1 = np.random.randn(n_h,n_x)*0.01 # n_h x n_x
    b1 = np.zeros((n_h,1))
    
    W2 = np.random.randn(n_y,n_h)*0.01 # n_y x n_h
    b2=np.zeros((n_y,1))
    
    parameters={
        "W1":W1,"b1":b1,"W2":W2,"b2":b2
    }
    return parameters
```

```python
#checking
n_x , n_h , n_y = initialize_parameters_test_case()

parameters = initialize_parameters(n_x,n_h,n_y)

print("W1"+str(parameters["W1"]))
print("b1"+str(parameters["b1"]))
print("W2"+str(parameters["W2"]))
print("b2"+str(parameters["b2"]))
```

```python
#forward propagation
def forward_propagation(X,parameters):
    # X is 2 x 400 --> (n_x , m)
    W1 = parameters["W1"] #n_h x n_x 
    b1 = parameters["b1"] #n_h x 1
    W2 = parameters["W2"] #n_y x n_h
    b2 = parameters["b2"] #n_y x 1
    
    Z1 = np.dot(W1,X) + b1 #n_h x m
    A1 = np.tanh(Z1)
    
    Z2 = np.dot(W2,A1) + b2 #n_y x m
    A2 = sigmoid(Z2)
    
    assert(A2.shape == (1,X.shape[1]))
    
    cache = {
        "Z1":Z1 , "A1":A1 , "Z2":Z2 , "A2":A2
    }
    return A2,cache
    
```

```python
#checking
X_assess, parameters = forward_propagation_test_case()
A2, cache = forward_propagation(X_assess, parameters)

# Note: we use the mean here just to make sure that your output matches ours. 
print(np.mean(cache['Z1']) ,np.mean(cache['A1']),np.mean(cache['Z2']),np.mean(cache['A2']))
```

```python
#computing cost (cross entropy)
def compute_cost(A2,Y,parameters):
    m = Y.shape[1]
    inter = np.multiply(np.log(A2),Y) + np.multiply(np.log(1-A2),(1-Y))
    cost  = -1/m*np.sum(inter)
    cost = np.squeeze(cost) #e.g. turns [[17]] to 17
    assert(isinstance(cost,float))
    return cost
    
```

```python
A2 , Y_assess , parameters = compute_cost_test_case()
cost = compute_cost(A2,Y_assess,parameters)
print("cost"+str(cost))
```

```python
def backward_propagation(parameters , cache , X , Y ):
    m=X.shape[1]
    W1 = parameters["W1"]
    W2 = parameters["W2"] #n_y x n_h
    
    A1=cache["A1"] #n_h x m
    A2=cache["A2"]
    
    #And back propagation begins
    dZ2 = A2 - Y #n_y x m
    dW2 = 1/m*np.dot(dZ2,A1.T) #n_h x n_y
    db2 = 1/m*np.sum(dZ2,axis=1,keepdims=True)
    
    dZ1 = np.dot(W2.T,dZ2)*(1-np.power(A1,2))
    dW1 = 1/m*np.dot(dZ1,X.T)
    db1 = 1/m*np.sum(dZ1 , axis=1,keepdims=True)
    
    grads={
        "dW1":dW1,
        "db1":db1,
        "dW2":dW2,
        "db2":db2
    }
    return grads
```

```python
#checking
parameters , cache , X_assess , Y_assess = backward_propagation_test_case()
grads = backward_propagation(parameters , cache , X_assess , Y_assess)
print("dW1="+str(grads["dW1"]))
print("db1="+str(grads["db1"]))
print("dW2="+str(grads["dW2"]))
print("db2="+str(grads["db2"]))
```

```python
#gradient descent
def update_parameters(parameters , grads , learning_rate=1.2):
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    b1 = parameters["b1"]
    b2 = parameters["b2"]
    
    dW1 = grads["dW1"]
    dW2 = grads["dW2"]
    db1 = grads["db1"]
    db2 = grads["db2"]
    
    W1 = W1 - learning_rate*dW1
    W2 = W2 - learning_rate*dW2
    b1 = b1 - learning_rate*db1
    b2 = b2 - learning_rate*db2
    
    parameters = {
        "W1":W1,
        "W2":W2,
        "b2":b2,
        "b1":b1
    }
    return parameters
```

```python
# checking
parameters , grads = update_parameters_test_case()
parameters = update_parameters(parameters,grads)

print("W1="+str(parameters["W1"]))
print("b1="+str(parameters["b1"]))
print("W2="+str(parameters["W2"]))
print("b2="+str(parameters["b2"]))
```

```python
def nn_model(X,Y,n_h,num_iter=10000,print_cost=False):
        np.random.seed(3)
        n_x = layer_sizes(X,Y)[0]
        n_y = layer_sizes(X,Y)[2]
        
        parameters = initialize_parameters(n_x,n_h,n_y)
        W1 = parameters["W1"]
        b1 = parameters["b1"]
        W2 = parameters["W2"]
        b2 = parameters["b2"]
        
        for i in range(0,num_iter):
            A2,cache = forward_propagation(X,parameters)
            
            cost = compute_cost(A2,Y,parameters)
            
            grads = backward_propagation(parameters,cache,X,Y)
            
            parameters = update_parameters(parameters,grads)
            
            if print_cost and i%100 ==0:
                print("Cost after iteration %i:%f " % (i,cost))
        return parameters
```
